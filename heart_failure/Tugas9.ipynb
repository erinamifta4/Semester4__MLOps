```python
import pandas as pd
from pycaret.datasets import get_data
from pycaret.classification import *
```

```python
# Memuat data dari lokasi file
data = pd.read_csv('C:/Users/USER/Semester 4_MLOps/heart_failure/heart_failure_clinical_records_dataset.csv')
data
```

```python
# Setup klasifikasi
s = setup(data, target='DEATH_EVENT', session_id=123)

# Membuat model klasifikasi
best = compare_models()

# Melakukan tuning pada model terbaik
tuned_model = tune_model(best)

# Menampilkan evaluasi model
evaluate_model(tuned_model)

# Menyelesaikan model
final_model = finalize_model(tuned_model)

# Menyimpan model ke file .pkl
save_model(final_model, 'heart_failure_classification_model')
```
