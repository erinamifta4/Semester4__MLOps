```python
import pandas as pd
from pycaret.classification import *

# Memuat data dari lokasi file
data = pd.read_csv('C:/Users/USER/Semester 4_MLOps/migraine/migraine_data.csv')

# Menghapus kolom 'UserID' dan 'Date'
data = data.drop(['DPF'], axis=1)

data
```

```python
# Setup klasifikasi
s = setup(data, target='Type', session_id=123)

# Membuat model klasifikasi
best = compare_models()

# Melakukan tuning pada model terbaik
model = tune_model(best)

# Menampilkan evaluasi model
evaluate_model(model)

# Menyelesaikan model
finalize_model(model)

# Menyimpan model ke file .pkl
save_model(model, 'migraine_model')
```
