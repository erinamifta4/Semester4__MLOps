```python
import pandas as pd
from pycaret.classification import *

# Memuat data dari lokasi file
data = pd.read_csv('C:/Users/USER/Semester 4_MLOps/kelayakan_ukt/klasifikasi_mahasiswa.csv')
data
```

```python
# Setup klasifikasi
s = setup(data, target='Kelayakan Keringanan UKT', session_id=123)

# Membuat model klasifikasi
best = compare_models()

# Melakukan tuning pada model terbaik
tuned_best = tune_model(best)

# Menampilkan evaluasi model
evaluate_model(tuned_best)

# Menyelesaikan model
final_model = finalize_model(tuned_best)

# Menyimpan model ke file .pkl
save_model(final_model, 'kelayakan_ukt')
```
