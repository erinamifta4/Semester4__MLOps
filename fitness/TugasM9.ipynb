```python
import pandas as pd
from pycaret.regression import *
```

```python
# Memuat data dari lokasi file
data = pd.read_csv('C:/Users/USER/Semester 4_MLOps/fitness/fitness.csv')

# Menghapus kolom 'UserID' dan 'Date'
data = data.drop(['UserID', 'Date'], axis=1)

data
```

```python
# Setup regression
s = setup(data, target='Calories_Burned', session_id=123)

# Membuat model regresi
best = compare_models()

# Melakukan tuning pada model terbaik
model = tune_model(best)

# Menampilkan evaluasi model
evaluate_model(model)

# Menyelesaikan model
finalize_model(model)

# Menyimpan model ke file .pkl
save_model(model, 'calories_regression_model')
```
